```python
# transpose labels

df_modelling_std_01_labels["GEN_Class2"]="NON-GEN"
df_modelling_std_01_labels["GEN_Class3"]="NON-GEN"
```

```python
for i in range(len(df_modelling_std_01_labels)):
    if df_modelling_std_01_labels.loc[i,"Class_2_status"] == "GEN":
        df_modelling_std_01_labels.loc[i,"GEN_Class2"] = "GEN"
    if df_modelling_std_01_labels.loc[i,"Class_3_status"] == "MargGen":
        df_modelling_std_01_labels.loc[i,"GEN_Class3"] = "MargGen"
    if df_modelling_std_01_labels.loc[i,"Class_3_status"] == "MainGen":
        df_modelling_std_01_labels.loc[i,"GEN_Class3"] = "MainGen"
    if df_modelling_std_01_labels.loc[i,"Class_3_status"] == "SupGen":
        df_modelling_std_01_labels.loc[i,"GEN_Class3"] = "SupGen"
    else:
        continue
```
