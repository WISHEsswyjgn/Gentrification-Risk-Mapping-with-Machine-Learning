## Analysing Counts of LSOAs According to their Neighbourhoo States

```python
# group lsoas according to their local authority

df_LA_counts = df_score.groupby(["Local_authority_name"]).size().reset_index()
```

```python
# checking the number of lsoas that each local authority has respectively

df_LA_counts = pd.DataFrame(df_LA_counts)
df_LA_counts = df_LA_counts.rename(columns={0:"LSOA_count"})
```

```python
df_LA_counts.head()
```

```python
# now group lsoas by their local authority and then their neighbourhood states

df_LA_break = df_score.groupby(["Local_authority_name","status"]).size().reset_index()
```

```python
# convert to dataframe and rename the data feature

df_LA_break = pd.DataFrame(df_LA_break)
df_LA_break = df_LA_break.rename(columns={0:"count"})
```

```python
df_LA_break.head()
```

```python
# merge the 2 dataframes

df_LA_counts = df_LA_counts.merge(df_LA_break, left_on = "Local_authority_name", right_on = "Local_authority_name", how = "left")
```

```python
df_LA_counts.head()
```

```python
# compute the share of LSOA neighbourhood states in each local authority (in % terms)

df_LA_counts["count_share"] = df_LA_counts["count"] / df_LA_counts["LSOA_count"] * 100
df_LA_counts["count_share"] = df_LA_counts["count_share"].round(2)
```

```python
df_LA_counts.head()
```

```python
# extract info for only the ascending states

df_LA_counts_asc = df_LA_counts.loc[df_LA_counts["status"] == "ASC"]
```

```python
# extract the 5 local authorities with the largest share of ascending lsoas

df_LA_counts_asc.nlargest(5, "count_share")
```

```python
# extract info for only the descending states

df_LA_counts_dec = df_LA_counts.loc[df_LA_counts["status"] == "DEC"]
```

```python
# extract the 5 local authorities with the largest share of descending lsoas

df_LA_counts_dec.nlargest(5, "count_share")
```
